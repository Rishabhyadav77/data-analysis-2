```python
# Load the Superstore dataset
df = pd.read_csv('SampleSuperstore.csv')
df.head()
```

```python
# Check the data types of each column
print(df.info())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python

# Data Visualization
# Visualize the sales distribution
sns.histplot(data=df, x='Sales')
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Count')
plt.show()
```

```python
# Visualize the relationship between Sales and Profit
sns.scatterplot(data=df, x='Sales', y='Profit')
plt.title('Sales vs Profit')
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.show()
```

```python
# Data Accessibility
# Filter the dataset to retrieve relevant information
filtered_data = df[(df['Region'] == 'West') & (df['Category'] == 'Office Supplies')]
```

```python
# Display the filtered data
print(filtered_data.head())
```

```python
# Operational Efficiency Enhancements
# Analyze inventory levels and identify potential stockouts
stockouts_data = df[df['Quantity'] <= 0]


```

```python
# Print the number of stockouts
print("Number of Stockouts:", stockouts_data)
```

```python
# Informed Decision-Making
# Calculate total sales by region
sales_by_region = df.groupby('Region')['Sales'].sum()

# Display the total sales by region
print(sales_by_region)
```

```python
# Streamlined Reporting Mechanisms
# Create a dashboard with key metrics
dashboard_data = df[['Region', 'Category', 'Sales', 'Profit']]
dashboard_data['Profit Margin'] = (dashboard_data['Profit'] / dashboard_data['Sales']) * 100

# Display the dashboard data
print(dashboard_data.head())
```
